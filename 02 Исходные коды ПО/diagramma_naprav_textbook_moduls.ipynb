```python
import numpy as np
import matplotlib.pyplot as plt

# Значения theta (в радианах)
theta = np.linspace(0, 2 * np.pi, 360)

# Значения degrees (в градусах)
degrees = np.linspace(0, 360, 360)

# Построение диаграммы направленности антенны
plt.polar(theta, degrees)

# Настройка осей и меток
plt.ylim(0, 360)
plt.xticks(np.linspace(0, 2 * np.pi, 12), ['0°', '30°', '60°', '90°', '120°', '150°', '180°', '210°', '240°', '270°', '300°', '330°'])

# Добавление заголовка
plt.title("Диаграмма направленности антенны")

# Отображение диаграммы
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema

plt.figure(figsize=(12, 6), dpi=200)

# Углы (в градусах)
angles = np.arange(-180, 190, 10)

# Усиление антенны (в дБи)
gains = [0, 3, 5, 3, 0, 3, 7, 3, 0, 5, 8, 5, 0, 6, 9, 6, 0, 9, 14, 9, 0, 6, 9,  0, 5, 8, 5, 0, 3, 7, 3, 0, 3, 5, 3, 0, 0]

# Построение диаграммы направленности антенны в прямоугольных координатах
plt.plot(angles, gains)

# Находим индексы локальных максимумов
local_maxima_indices = argrelextrema(np.array(gains), np.greater)

# Извлекаем углы и усиления для локальных максимумов
local_maxima_angles = angles[local_maxima_indices]
local_maxima_gains = np.array(gains)[local_maxima_indices]

# Построение точек локальных максимумов на графике
plt.scatter(local_maxima_angles, local_maxima_gains, color='red', label='Local Maxima')

# Аннотация для каждого локального максимума
for angle, gain in zip(local_maxima_angles, local_maxima_gains):
    plt.annotate(f'({angle}, {gain})', xy=(angle, gain), xytext=(angle - 20, gain + 1), color='red')

# Настройка осей и меток
plt.xlabel("Угол (градусы)")
plt.ylabel("Усиление (дБи)")
plt.xlim(-180, 180)
plt.ylim(0, max(gains) + 2)

# Добавление заголовка
plt.title("Диаграмма направленности антенны")

# Отображение сетки
plt.grid()

# Добавление легенды
plt.legend()

# Отображение диаграммы
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema

plt.figure(figsize=(6, 6), dpi=200)

# Углы (в градусах)
angles = np.deg2rad(np.arange(-180, 190, 10))

# Усиление антенны (в дБи)
gains = [0, 3, 5, 3, 0, 3, 7, 3, 0, 5, 8, 5, 0, 6, 9, 6, 0, 9, 14, 9, 0, 6, 9, 9, 0, 5, 8, 5, 0, 3, 7, 3, 0, 3, 5, 3, 0]

# Построение диаграммы направленности антенны в полярных координатах
plt.polar(angles, gains)

# Находим индексы локальных максимумов
local_maxima_indices = argrelextrema(np.array(gains), np.greater)

# Извлекаем углы и усиления для локальных максимумов
local_maxima_angles = angles[local_maxima_indices]
local_maxima_gains = np.array(gains)[local_maxima_indices]

# Построение точек локальных максимумов на графике
plt.scatter(local_maxima_angles, local_maxima_gains, color='red', label='Local Maxima')

# Аннотация для каждого локального максимума
for angle, gain in zip(local_maxima_angles, local_maxima_gains):
    plt.annotate(f'({np.rad2deg(angle):.0f}, {gain})', xy=(angle, gain), xytext=(angle + 0.1, gain + 1), color='red')

# Добавление заголовка
plt.title("Диаграмма направленности антенны")


# Добавление легенды
plt.legend()

# Отображение диаграммы
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema
from mpl_toolkits.mplot3d import Axes3D

plt.figure(figsize=(8, 8), dpi=200)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Углы (в градусах)
angles = np.deg2rad(np.arange(-180, 190, 10))

# Усиление антенны (в дБи)
gains = [0, 3, 5, 3, 0, 3, 7, 3, 0, 5, 8, 5, 0, 6, 9, 6, 0, 9, 14, 9, 0, 6, 9, 9, 0, 5, 8, 5, 0, 3, 7, 3, 0, 3, 5, 3, 0]

# Построение 3D графика направленности антенны
ax.plot(angles, gains, zs=0, zdir='z', label='Antenna Directivity')

# Находим индексы локальных максимумов
local_maxima_indices = argrelextrema(np.array(gains), np.greater)

# Извлекаем углы и усиления для локальных максимумов
local_maxima_angles = angles[local_maxima_indices]
local_maxima_gains = np.array(gains)[local_maxima_indices]

# Построение точек локальных максимумов на графике
ax.scatter(local_maxima_angles, local_maxima_gains, np.zeros_like(local_maxima_gains), c='red', label='Local Maxima')

# Аннотация для каждого локального максимума
for angle, gain in zip(local_maxima_angles, local_maxima_gains):
    ax.text(angle, gain, 0, f'({np.rad2deg(angle):.0f}, {gain})', color='red')

# Настройка осей и меток
ax.set_xlabel("Угол (радианы)")
ax.set_ylabel("Усиление (дБи)")
ax.set_zlabel("Высота")
ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(0, max(gains) + 2)
ax.set_zlim(0, max(gains) + 2)

# Добавление заголовка
ax.set_title("3D Диаграмма направленности антенны")

# Добавление легенды
ax.legend()

# Отображение графика
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LightSource

# Create data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Increase the size of the plot window
fig = plt.figure(figsize=(10, 8))  # Set width=10 inches, height=8 inches
ax = fig.add_subplot(111, projection='3d')

# Create light source object
light_source = LightSource(azdeg=315, altdeg=45)

# Calculate hillshading
rgb = light_source.shade(Z, cmap=plt.cm.cool)

# Plot the surface with hillshading
surf = ax.plot_surface(X, Y, Z, facecolors=rgb, rstride=1, cstride=1)

# Add a color bar
fig.colorbar(surf)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Custom Hillshading in 3D Surface Plot')

# Show the plot
plt.show()
```
